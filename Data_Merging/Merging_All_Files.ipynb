```python
import pandas as pd
import numpy as np
```

```python
read_06 = pd.read_csv('../../Personal_Project/Source_File/Data_Merging/fa_2006_list.csv')
read_07 = pd.read_csv('../../Personal_Project/Source_File/Data_Merging/fa_2007_list.csv')
read_08 = pd.read_csv('../../Personal_Project/Source_File/Data_Merging/fa_2008_list.csv')
read_09 = pd.read_csv('../../Personal_Project/Source_File/Data_Merging/fa_2009_list.csv')
read_10 = pd.read_csv('../../Personal_Project/Source_File/Data_Merging/fa_2010_list.csv')
read_11 = pd.read_csv('../../Personal_Project/Source_File/Data_Merging/fa_2011_list.csv')
read_12 = pd.read_csv('../../Personal_Project/Source_File/Data_Merging/fa_2012_list.csv')
read_13 = pd.read_csv('../../Personal_Project/Source_File/Data_Merging/fa_2013_list.csv')
read_14 = pd.read_csv('../../Personal_Project/Source_File/Data_Merging/fa_2014_list.csv')
read_15 = pd.read_csv('../../Personal_Project/Source_File/Data_Merging/fa_2015_list.csv')
read_16 = pd.read_csv('../../Personal_Project/Source_File/Data_Merging/fa_2016_list.csv')
read_17 = pd.read_csv('../../Personal_Project/Source_File/Data_Merging/fa_2017_list.csv')
```

```python
merge1 = pd.concat([read_06, read_07])
merge2 = pd.concat([merge1, read_08])
merge3 = pd.concat([merge2, read_09])
merge4 = pd.concat([merge3, read_10])
merge5 = pd.concat([merge4, read_11])
merge6 = pd.concat([merge5, read_12])
merge7 = pd.concat([merge6, read_13])
merge8 = pd.concat([merge7, read_14])
merge9 = pd.concat([merge8, read_15])
merge10 = pd.concat([merge9, read_16])
merge = pd.concat([merge10, read_17])
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

merge_1 = merge.set_index('Name', inplace=False)
merge_2 = merge_1.reset_index()

merge = merge_2
```
