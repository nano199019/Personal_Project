```python
import pandas as pd
import numpy as np
import re
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
read = pd.read_csv('../../csv/FA_LIST_2008.csv')
read_dash_b = pd.read_csv('../csv/2008/dash_b.csv')
read_std_b = pd.read_csv('../csv/2008/std_b.csv')
read_adv_b = pd.read_csv('../csv/2008/adv_b.csv')
read_wpa_b = pd.read_csv('../csv/2008/wpa_b.csv')
read_dash_p = pd.read_csv('../csv/2008/dash_p.csv')
read_std_p = pd.read_csv('../csv/2008/std_p.csv')
read_adv_p = pd.read_csv('../csv/2008/adv_p.csv')
read_wpa_p = pd.read_csv('../csv/2008/wpa_p.csv')
```

```python
copy = read.copy()
rdb = read_dash_b.copy()
rsb = read_std_b.copy()
rab = read_adv_b.copy()
rwb = read_wpa_b.copy()
rdp = read_dash_p.copy()
rsp = read_std_p.copy()
rap = read_adv_p.copy()
rwp = read_wpa_p.copy()
```

## Batters

```python
merge1 = rdb.merge(rsb, how='inner')
merge2 = merge1.merge(rab, how='inner')
merge_b = merge2.merge(rwb, how='inner')
```

```python
to_list_b = merge_b.columns.tolist()

to_list_b = ['Name', 'Team', 'G', 'AB', 'PA', 'H', '1B', '2B', '3B', 'HR', 'R', 'RBI', 'BB', 'IBB', 'SO', 'BB%', 'K%', 'BB/K', 'BABIP', 'HBP', 'SF', 'SH', 'GDP', 'SB', 'CS', 'Def', 'ISO', 'wOBA', 'wRAA', 'AVG', 'OBP', 'SLG', 'OPS', 'WAR', 'wRC+', 'WPA']
merge_b = merge_b[to_list_b]
```

```python
merge_b_copy = merge_b.copy()
```

```python
bb_list = list(merge_b_copy['BB%'])
k_list = list(merge_b_copy['K%'])

bb = [float(re.sub(r' %', '', x)) for x in list(bb_list)]
k = [float(re.sub(r' %', '', x)) for x in list(k_list)]
```

```python
merge_b['BB%'] = bb
merge_b['K%'] = k
```

```python
merge_b = merge_b[merge_b['PA'] > 70]

print(merge_b.shape)
merge_b
```

```python
copy_b_name = copy['Name']
merge_b_name = merge_b['Name']

names_b = pd.DataFrame({'name_1' : merge_b_name})
names_2_b = pd.DataFrame({'name_2' : copy_b_name})

names_b.insert(1, 'names_2', names_2_b)

names_b
```

```python
overlap_names = pd.Series(list(set(merge_b_name).intersection(set(copy_b_name))))
overlap_names = pd.Series(np.intersect1d(merge_b_name, copy_b_name))
overlap_names = pd.Series(np.intersect1d(merge_b_name.values, copy_b_name.values))

overlap_names = pd.DataFrame({'Name' : overlap_names})

print(len(overlap_names))
overlap_names
```

```python
fa_2008_list_b = pd.merge(overlap_names, merge_b, how='inner', on='Name')

print(fa_2008_list_b.shape)
fa_2008_list_b
```

## Pitchers

```python
merge3 = rdp.merge(rsp, how='inner')
merge4 = merge3.merge(rap, how='inner')
merge_pit = merge4.merge(rwp, how='inner')
```

```python
to_list = merge_pit.columns.tolist()

to_list = ['Name', 'Team', 'W', 'L', 'G', 'GS', 'CG', 'IP', 'H', 'HR', 'BB', 'SO', 'R', 'SV', 'BS', 'BK', 'ER', 'HBP', 'HLD', 'IBB', 'ShO', 'TBF', 'WP', 'HR/9', 'BB/9', 'K/9', 'K/BB', 'K-BB%', 'GB%', 'HR/FB', 'LOB%', 'ERA', 'BABIP', 'FIP', 'WHIP', 'WPA', 'WAR']
merged_data = merge_pit[to_list]
```

```python
merged_data = merged_data[merged_data['IP'] > 5]

print(merged_data.shape)
merged_data
```

```python
k_bb_list = merged_data['K-BB%']
gb_list = merged_data['GB%']
hr_fb_list = merged_data['HR/FB']
lob_list = merged_data['LOB%']

k_bb = [float(re.sub(r' %', '', x)) for x in list(k_bb_list)]
gb = [float(re.sub(r' %', '', x)) for x in list(gb_list)]
hr_fb = [float(re.sub(r' %', '', x)) for x in list(hr_fb_list)]
lob = [float(re.sub(r' %', '', x)) for x in list(lob_list)]
```

```python
merged_data['K-BB%'] = k_bb
merged_data['GB%'] = gb
merged_data['HR/FB'] = hr_fb
merged_data['LOB%'] = lob
```

```python
copy_name = copy['Name']
merge_pit_name = merged_data['Name']

names_p = pd.DataFrame({'name_1' : merge_pit_name})
names_p2 = pd.DataFrame({'name_2' : copy_name})

names_p.insert(1, 'names_2', names_p2)

names_p
```

```python
overlap_names2 = pd.Series(list(set(merge_pit_name).intersection(set(copy_name))))
overlap_names2 = pd.Series(np.intersect1d(merge_pit_name, copy_name))
overlap_names2 = pd.Series(np.intersect1d(merge_pit_name.values, copy_name.values))

overlap_names2 = pd.DataFrame({'Name' : overlap_names2})

print(len(overlap_names2))
overlap_names2
```

```python
fa_2008_list_p = pd.merge(overlap_names2, merged_data, how='inner', on='Name')

print(fa_2008_list_p.shape)
fa_2008_list_p
```
