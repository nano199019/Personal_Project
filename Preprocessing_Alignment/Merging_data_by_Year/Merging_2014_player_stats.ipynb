```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
read_bat = pd.read_csv('../csv/2014/2014_Merging_batters_stats.csv')
read_pic = pd.read_csv('../csv/2014/2014_Merging_pitchers_stats.csv')
```

```python
rb = read_bat.copy()
rp = read_pic.copy()
```

```python
merged_data_2014 = rb.merge(rp, how='outer', on=('Name', 'Team', 'G', 'WAR', 'WPA'), suffixes=('_b', '_p')).fillna(0)
```
