```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
read_2017_std = pd.read_csv('../csv/Merging_2017_STD_Pichers.csv')
read_2017_dash = pd.read_csv('../csv/Merging_2017_DASH_Pichers.csv')
read_2017_adv = pd.read_csv('../csv/Merging_2017_ADV_Pichers.csv')
```

```python
copy_std = read_2017_std.copy()
copy_dash = read_2017_dash.copy()
copy_adv = read_2017_adv.copy()
```

```python
merge1 = pd.merge(copy_std, copy_dash, on=['Name', 'Team'])
merged_data = pd.merge(merge1, copy_adv, on=['Name', 'Team'])

print(merged_data.shape)
merged_data
```

```python
to_list = merged_data.columns.tolist()
print(to_list)
```

```python
to_list = ['Name', 'Team', 'W', 'L', 'G', 'GS', 'CG', 'IP', 'H', 'HR', 'BB', 'SO', 'R', 'SV', 'BS', 'BK', 'ER', 'HBP', 'HLD', 'IBB', 'ShO', 'TBF', 'WP', 'HR/9', 'BB/9', 'K/9', 'K/BB', 'K-BB%', 'GB%', 'HR/FB', 'LOB%', 'ERA', 'BABIP', 'FIP', 'WHIP', 'WPA', 'WAR']
merged_data = merged_data[to_list]

print(merged_data.shape)
merged_data
```

```python
merged_data.to_csv('../csv/2017_Merging_pichers_stats.csv', index=None)
```
