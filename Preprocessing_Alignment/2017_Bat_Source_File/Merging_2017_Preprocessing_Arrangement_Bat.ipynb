```python
read_2017_std = pd.read_csv('../csv/Merging_2017_STD_bat.csv')
read_2017_dash = pd.read_csv('../csv/Merging_2017_DASH_bat.csv')
read_2017_adv = pd.read_csv('../csv/Merging_2017_AVD_bat.csv')
```
