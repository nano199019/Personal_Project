## 2017 Advanced Stats. Page1

```python
res = requests.get('http://www.fangraphs.com/leaders.aspx?pos=all&stats=bat&lg=all&qual=y&type=1&season=2017&month=0&season1=2017&ind=0&team=0&rost=0&age=0&filter=&players=0&page=1_50')
soup = BeautifulSoup(res.content, 'html.parser')

player_dict = {}
for player in soup.select('tbody > tr'):
    try:
        name = player.select('td > a')[0].text
    except:
        name = None
    try:
        team = player.select('td > a')[1].text
    except:
        team = None
    player_spec = []
    for i in player.select('td')[3:]:
        player_spec.append(i.text)
    player_dict[(name, team)] = player_spec
print(len(player_dict.keys()))
pprint(player_dict)
```

```python
key_list = list(player_dict.keys())
key_list = key_list[1:]
key_list
```

```python
name_list = []
team_list = []

for i in range(len(key_list)):
    name_list.append(key_list[i][0])

for i in range(len(key_list)):
    team_list.append(key_list[i][1])
    

print(name_list, '\n\n', team_list)
```

```python
name_list = pd.DataFrame({'Name' : name_list})
team_list = pd.DataFrame({'Team' : team_list})

name_list.insert(1, 'Team', team_list)
name_team_list = name_list
```

```python
print(len(name_team_list))
name_team_list
```

```python
name_team_list.loc[22, 'Team'] = 'Angels'
name_team_list.loc[32, 'Team'] = 'Mariners'

name_team_list
```

```python
real_stats = list(player_dict.values())
```

```python
real_stats = real_stats[1:]
print(real_stats)
```

```python
bb_k_rate = [float(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[3]]
ops_num = [float(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[7]]
wraa_num = [float(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[15]]
```

```python
temporary = pd.DataFrame({'BB/K' : bb_k_rate, 'OPS' : ops_num, 'wRAA' : wraa_num})
```

```python
temporary
```

```python
asdasd = temporary.copy()
```

```python
merged_data1 = pd.concat([name_team_list, asdasd], axis=1)
merged_data1
```
