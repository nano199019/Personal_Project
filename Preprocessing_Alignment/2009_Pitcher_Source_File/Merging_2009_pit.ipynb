```python
import pandas as pd
import numpy as np
import re
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
read_dash = pd.read_csv('../csv/2009/Merging_2009_DASH_pitcher.csv')
read_std = pd.read_csv('../csv/2009/Merging_2009_STD_pitcher.csv')
read_adv = pd.read_csv('../csv/2009/Merging_2009_ADV_pitcher.csv')
read_wpa = pd.read_csv('../csv/2009/Merging_2009_WPA_pitcher.csv')
```

```python
rd = read_dash.copy()
rs = read_std.copy()
ra = read_adv.copy()
rw = read_wpa.copy()
```

```python
merge1 = rd.merge(rs, how='inner')
merge2 = merge1.merge(ra, how='inner')
merge = merge2.merge(rw, how='inner')
```

```python
to_list = merge.columns.tolist()

to_list = ['Name', 'Team', 'W', 'L', 'G', 'GS', 'CG', 'IP', 'H', 'HR', 'BB', 'SO', 'R', 'SV', 'BS', 'BK', 'ER', 'HBP', 'HLD', 'IBB', 'ShO', 'TBF', 'WP', 'HR/9', 'BB/9', 'K/9', 'K/BB', 'K-BB%', 'GB%', 'HR/FB', 'LOB%', 'ERA', 'BABIP', 'FIP', 'WHIP', 'WPA', 'WAR']
merge = merge[to_list]

print(merge.shape)
merge
```

```python
copy = merge.copy()
```

```python
k_bb_list = merge['K-BB%']
gb_list = merge['GB%']
hr_fb_list = merge['HR/FB']
lob_list = merge['LOB%']
```

```python
k_bb = [float(re.sub(r' %', '', x)) for x in list(k_bb_list)]
gb = [float(re.sub(r' %', '', x)) for x in list(gb_list)]
hr_fb = [float(re.sub(r' %', '', x)) for x in list(hr_fb_list)]
lob = [float(re.sub(r' %', '', x)) for x in list(lob_list)]
```

```python
merge['K-BB%'] = k_bb
merge['GB%'] = gb
merge['HR/FB'] = hr_fb
merge['LOB%'] = lob
```

```python
print(merge.shape)
merge
```

```python
merge.loc[5, 'Team'] = 'Phillies'
merge.loc[34, 'Team'] = 'Twins'
merge.loc[59, 'Team'] = 'Dodgers'
merge.loc[63, 'Team'] = 'Tigers'
merge.loc[64, 'Team'] = 'Giants'
merge.loc[67, 'Team'] = 'Nationals'
```
