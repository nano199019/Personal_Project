```python
import pandas as pd
import numpy as np
import re
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
read_dash = pd.read_csv('../csv/2016/Merging_2016_DASH_bat.csv')
read_std = pd.read_csv('../csv/2016/Merging_2016_STD_bat.csv')
read_adv = pd.read_csv('../csv/2016/Merging_2016_ADV_bat.csv')
read_wpa = pd.read_csv('../csv/2016/Merging_2016_WPA_bat.csv')
```

```python
rd = read_dash.copy()
rs = read_std.copy()
ra = read_adv.copy()
rw = read_wpa.copy()
```

```python
merge1 = rd.merge(rs, how='inner')#, on=('Name', 'Team'))
merge2 = merge1.merge(ra, how='inner')#, on=('Name', 'Team'))
merge = merge2.merge(rw, how='inner')#, on=('Name', 'Team'))

print(merge.shape)
merge
```

```python
to_list = merge.columns.tolist()
print(to_list)
```

```python
to_list = ['Name', 'Team', 'G', 'AB', 'PA', 'H', '1B', '2B', '3B', 'HR', 'R', 'RBI', 'BB', 'IBB', 'SO', 'BB%', 'K%', 'BB/K', 'BABIP', 'HBP', 'SF', 'SH', 'GDP', 'SB', 'CS', 'Def', 'ISO', 'wOBA', 'wRAA', 'AVG', 'OBP', 'SLG', 'OPS', 'WAR', 'wRC+', 'WPA']
merge = merge[to_list]

print(merge.shape)
merge
```

```python
copy = merge.copy()
```

```python
bb_list = list(copy['BB%'])
k_list = list(copy['K%'])
```

```python
bb = [float(re.sub(r' %', '', x)) for x in list(bb_list)]
k = [float(re.sub(r' %', '', x)) for x in list(k_list)]
```

```python
merge['BB%'] = bb
merge['K%'] = k

merge
```
