# 2017 DashBoard Stats. Page1

```python
res = requests.get('http://www.fangraphs.com/leaders.aspx?pos=all&stats=bat&lg=all&qual=y&type=8&season=2017&month=0&season1=2017&ind=0&team=0&rost=0&age=0&filter=&players=0&page=1_50')
soup = BeautifulSoup(res.content, 'html.parser')

player_dict = {}
for player in soup.select('tbody > tr'):
    try:
        name = player.select('td > a')[0].text
    except:
        name = None
    try:
        team = player.select('td > a')[1].text
    except:
        team = None
    player_spec = []
    for i in player.select('td')[3:]:
        player_spec.append(i.text)
    player_dict[(name, team)] = player_spec
print(len(player_dict.keys()))
pprint(player_dict)
```

```python
key_list = list(player_dict.keys())
key_list = key_list[1:]
key_list
```

```python
name_list = []
team_list = []

for i in range(len(key_list)):
    name_list.append(key_list[i][0])

for i in range(len(key_list)):
    team_list.append(key_list[i][1])
    

print(name_list, '\n\n', team_list)
```

```python
name_list = pd.DataFrame({'Name' : name_list})
team_list = pd.DataFrame({'Team' : team_list})

name_list.insert(1, 'Team', team_list)
name_team_list = name_list
```

```python
print(len(name_team_list))
name_team_list
```

```python
name_team_list.loc[16, 'Team'] = 'Angels'
name_team_list.loc[45, 'Team'] = 'Orioles'

name_team_list
```

```python
real_stats = list(player_dict.values())
```

```python
real_stats = real_stats[1:]
print(real_stats)
```

```python
bb_rate = [float(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[6]]
k_rate = [float(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[7]]
iso_num = [float(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[8]]
babip_num = [float(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[9]]
obp_num = [float(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[11]]
slg_num = [float(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[12]]
woba_num = [float(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[13]]
wrc_plus = [int(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[14]]
Def_num = [float(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[17]]
war_num = [float(re.sub(r' %', '', x)) for x in list(zip(*real_stats))[18]]
```

```python
temporary = pd.DataFrame({'BB%' : bb_rate, 'K%' : k_rate, 'ISO' : iso_num, 'BABIP' : babip_num, 'OBP' : obp_num, 'SLG' : slg_num, 'wOBA' : woba_num, 'wRC+' : wrc_plus, 'Def' : Def_num, 'WAR' : war_num})
```

```python
temporary
```

```python
asdasd = temporary.copy()
```

```python
merged_data1 = pd.concat([name_team_list, asdasd], axis=1)
merged_data1
```
