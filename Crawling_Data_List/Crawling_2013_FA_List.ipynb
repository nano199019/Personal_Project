### * Crawling the 'Name and Salary' part & Preprocessing

### 1. Extracting Name info

```python
driver = webdriver.PhantomJS('C:\\Users\\Jwon\\Downloads\\phantomjs-2.1.1-windows\\bin\\phantomjs.exe')
driver.get('http://www.spotrac.com/mlb/free-agents/2013/')

contents = driver.find_elements_by_css_selector("#main > div.teams > table > tbody > tr > td.player")

Name_list = [x.text for x in contents]
Name_data = pd.DataFrame({'Name' : Name_list})

driver.quit()
```

### 2. Extracting Contract Yrs info

```python
driver = webdriver.PhantomJS('C:\\Users\\Jwon\\Downloads\\phantomjs-2.1.1-windows\\bin\\phantomjs.exe')
driver.get('http://www.spotrac.com/mlb/free-agents/2013/')

Contract_Yrs_contents = driver.find_elements_by_css_selector('#main > div.teams > table > tbody > tr > td:nth-child(6)')

Contract_Yrs_list = [i.text for i in Contract_Yrs_contents]
Contract_Yrs_data = pd.DataFrame({'Contract Yrs' : Contract_Yrs_list})

driver.quit()
```

```python
Contract_Yrs_temp = Contract_Yrs_data[:140]
Contract_Yrs_temp
```

### 3. Extracting Salary Info

```python
driver = webdriver.PhantomJS('C:\\Users\\Jwon\\Downloads\\phantomjs-2.1.1-windows\\bin\\phantomjs.exe')
driver.get('http://www.spotrac.com/mlb/free-agents/2013/')

Salary_contents = driver.find_elements_by_css_selector('#main > div.teams > table > tbody > tr > td:nth-child(7)')

Salary_list = [i.text for i in Salary_contents]
Salary_data = pd.DataFrame({'Salary' : Salary_list})

driver.quit()
```

```python
Salary_temp = Salary_data[:140]
Salary_temp
```

### * Change Types 'Str' To 'Numeric' for 'Contract Yrs' & 'Salary' Columns

```python
def tonumeric(ser):
    return pd.to_numeric(ser.str.replace(',', '').str.replace('$', ''))
```

```python
tonumeric(Contract_Yrs_temp['Contract Yrs'])

Contract_Yrs = tonumeric(Contract_Yrs_temp['Contract Yrs'])
Contract_Yrs = pd.DataFrame({'Contract Yrs' : Contract_Yrs})

Contract_Yrs
```

```python
tonumeric(Salary_temp['Salary'])

Salary = tonumeric(Salary_temp['Salary'])
Salary = pd.DataFrame({'Total Salary($)' : Salary})

Salary
```

### * Merging 'Name', 'Contract Yrs' and 'Salary' Columns

```python
Names.insert(1, 'Total Salary($)', Salary)
Names.insert(1, 'Contract Yrs', Contract_Yrs)
Names
```

```python
Names.columns
```

### * Add New Column 'Average Salary'

```python
copy = Names.copy()

copy['Avg Salary($)'] = (copy['Total Salary($)'] / copy['Contract Yrs'])
Names = copy

Names
```

 ### 2013 MLB FA list (Except. Salary)

### Data Preprocessing

### * Overlapped Name extract

```python
names_1 = fa_2013_mlb['Name']
names_2 = Names['Name']

names = pd.DataFrame({'names_1' : names_1})

names_2 = pd.DataFrame({'names_2' : names_2})

names.insert(1, 'names_2', names_2)
names
```

```python
overlap_names = pd.Series(list(set(names_1).intersection(set(names_2))))
overlap_names = pd.Series(np.intersect1d(names_1, names_2))
overlap_names = pd.Series(np.intersect1d(names_1.values, names_2.values))

overlap_names = pd.DataFrame({'Name' : overlap_names})

print(len(overlap_names))
overlap_names
```

```python
fa_2013_list_temp = pd.merge(overlap_names, fa_2013_mlb, how='inner', on='Name')
fa_2013_list = pd.merge(fa_2013_list_temp, Names, how='inner', on='Name')
fa_2013_list
```
