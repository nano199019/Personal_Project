### 2. Extracting Contract Yrs

### 3. Extracting Salary info

```python
driver = webdriver.PhantomJS('C:\\Users\\Jwon\\Downloads\\phantomjs-2.1.1-windows\\bin\\phantomjs.exe')
driver.get('http://www.espn.com/mlb/freeagents/_/year/2010')

Contract_Yrs_contents = driver.find_elements_by_css_selector('#my-players-table > div.mod-container.mod-table.mod-no-header > div.mod-content \
> table > tbody > tr > td:nth-child(9)')

Salary_list = [i.text for i in Contract_Yrs_contents if i.text != 'DOLLARS']
Salary_data = pd.DataFrame({'Salary' : Salary_list})

driver.quit()
```

```python
Salary_temp = Salary_data
Salary_temp
```

### * Change Types 'Str' To 'Numeric' for 'Contract Yrs' & 'Salary' Columns

```python
def tonumeric(ser):
    return pd.to_numeric(ser.str.replace(',', '').str.replace('$', '').str.replace('Minor Lg', '').str.\
                        replace('--', ''))
```

```python
tonumeric(Contract_Yrs_temp['Contract Yrs'])

Contract_Yrs = tonumeric(Contract_Yrs_temp['Contract Yrs'])
Contract_Yrs = pd.DataFrame({'Contract Yrs' : Contract_Yrs})

Contract_Yrs
```

```python
tonumeric(Salary_temp['Salary'])

Salary = tonumeric(Salary_temp['Salary'])
Salary = pd.DataFrame({'Total Salary($)' : Salary})

Salary
```

### * Merging 'Name', 'Contract Yrs' and 'Salary' Columns

```python
Names.insert(1, 'Total Salary($)', Salary)
Names.insert(1, 'Contract Yrs', Contract_Yrs)
Names
```

### * Remove the Nan Values

```python
Names = Names.dropna()

print(Names.shape)
Names
```

### * Add New Column 'Average Salary'

```python
copy = Names.copy()

copy['Avg Salary($)'] = (copy['Total Salary($)'] / copy['Contract Yrs'])
Names = copy

Names
```

### * Extracting the Overlapped Name

```python
names_1 = fa_2010_mlb['Name']
names_2 = Names['Name']

names = pd.DataFrame({'names_1' : names_1})
names_2 = pd.DataFrame({'names_2' : names_2})

names.insert(1, 'names_2', names_2)
names
```

```python
overlap_names = pd.Series(list(set(names_1).intersection(set(names_2))))
overlap_names = pd.Series(np.intersect1d(names_1, names_2))
overlap_names = pd.Series(np.intersect1d(names_1.values, names_2.values))

overlap_names = pd.DataFrame({'Name' : overlap_names})

print(len(overlap_names))
overlap_names
```

```python
fa_2010_list_temp = pd.merge(overlap_names, fa_2010_mlb, how='inner', on='Name')
fa_2010_list = pd.merge(fa_2010_list_temp, Names, how='inner', on='Name')

fa_2010_list
```
